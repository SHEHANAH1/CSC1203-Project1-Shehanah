```python
import os

class Person:
    def __init__(self, name, age):
        self.name = name              # public
        self._age = age               # protected

    def display_info(self):
        print( "Name: ", self.name, ", Age:" , self._age, end=', ')



class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.__student_id = student_id   # private
        self.grade = grade

    # Getter (Encapsulation)
    def get_id(self):
        return self.__student_id

    # Setter with validation
    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.grade = grade
        else:
            print("Invalid grade!")

    def display_info(self):
        super().display_info()
        print("ID: ",self.__student_id, ",  Grade: ", self.grade)

def load_students():
    StudentsList = []
    if os.path.exists("C:\\Users\Administrator\\OneDrive - Tabuk Girls.txt"):
        with open("C:\\Users\Administrator\\OneDrive - Tabuk Girls.txt", "r") as f:
            for line in f:
                name, age, sid, grade = line.strip().split(",")
                StudentsList.append(Student(name, int(age), sid, float(grade)))

    return StudentsList



def save_students(students):
    f= open("C:\\Users\Administrator\\OneDrive - Tabuk Girls.txt", "w")
    for s in students:
        f.write(f"{s.name},{s._age},{s.get_id()},{s.grade}\n")
    f.close()    



def add_student(students):
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    sid = input("Enter student ID: ")
    grade = float(input("Enter grade: "))

    if grade <0 or  grade> 100:
        print("Invalid grade!")
    else:
        students.append(Student(name, age, sid, grade))    
        print("Student added successfully!")


def display_students(students):
    if not students:
        print("No students found.")
    for s in students:
        s.display_info()


def delete_student(students):
    sid = input("Enter ID to delete: ")
    for s in students:
        if s.get_id() == sid:
            students.remove(s)
            print("Deleted successfully!")
            return
        else:
            print("Student not found.")


def update_student(students):
    sid = input("Enter ID to update: ")
    for s in students:
        if s.get_id() == sid:
            new_grade = float(input("Enter new grade: "))
            s.set_grade(new_grade)
            print("Updated successfully!")
            return
        else:
            print("Student not found.")



students = load_students()

while True:
        print("\n===== Student System =====")
        print("1. Add Student")
        print("2. Display Students")
        print("3. Update Student")
        print("4. Delete Student")
        print("5. Exit")

        choice = input("Enter the Choose Number 1 - 5: ")

        if choice == "1":
            add_student(students)
        elif choice == "2":
            display_students(students)
        elif choice == "3":
            update_student(students)
        elif choice == "4":
            delete_student(students)
        elif choice == "5":
            save_students(students)
            print("Data saved.")
            break
        else:
            print("Invalid choice!")
```
